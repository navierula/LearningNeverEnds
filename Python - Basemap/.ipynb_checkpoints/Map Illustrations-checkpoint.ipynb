```python
# display contents in notebook
%matplotlib inline 
# import basemap library and matplot library
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
# set size of map
plt.figure(figsize=(12,6))
map = Basemap()
# draw a flat, basic map of the world
map.drawcoastlines()
plt.show()
```

```python
# draw a map with orthogonal projection
plt.figure(figsize=(12,6))
# change latitude and longitude to reflect
# Bangkok location 
map = Basemap(projection='ortho', 
              lat_0=13, lon_0=100)
# color the globe
map.drawmapboundary(fill_color='blue')

map.fillcontinents(color='green',lake_color='blue')
map.drawcoastlines()

plt.show()
```

```python
# different projections
plt.figure(figsize=(12,6))
map = Basemap(projection='cyl')

# flipping the colors!
map.drawmapboundary(fill_color='green')
map.fillcontinents(color='blue',lake_color='green')
map.drawcoastlines()

plt.show()
```
